## <u>Define the following terms:</u>

---

### Supervised Learning

machine learning where you have both the input and output and are trying to figure out the how

### Regression


statistical measurement that attempts to determine strength of relationship between a dependent variable and a series of changing variables


### Classification

task of approximating a mapping function (f) from input variables (X) to discrete output variables (y)

---



### Heuristic

using available data to solve machine and human problems.

### Descriptive Statistics

Summarizing and organizing your data so that it can be easily understood.

----

### What should you always begin with when constructing a predictive model?

A line of best fit using your descriptive statistics.

### How should one interpret the coefficients in a linear regression model?

it is the difference in the predicted value of the response variable for each one unit change in the predictor variable assuming all other predictor variables are held constant.
